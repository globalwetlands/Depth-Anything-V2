```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_path = '../data/diego/set1/csv/results_puffer_oyster.csv'
```

```python
# Read the CSV data
df = pd.read_csv(csv_path)
df.head()
```

```python
print(df.columns.tolist())
```

```python
# Display the data
print(df[['category_name', 'normalized_depth']])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='category_name', y='normalized_depth', data=df)
plt.title('Normalized Depth by Category')
plt.xlabel('Category Name')
plt.ylabel('Normalized Depth')
plt.show()
```

```python
# Calculate mean values for each category
means = df.groupby('category_name')['normalized_depth'].mean()

# Print the means
print("Mean Normalized Depth by Category:")
print(means)
```

```python
plt.figure(figsize=(10, 6))
sns.stripplot(x='category_name', y='normalized_depth', data=df, jitter=True)
plt.title('Normalized Depth by Category')
plt.xlabel('Category Name')
plt.ylabel('Normalized Depth')
plt.show()
```
